## Functions

#### column_name(df) 
Function to convert dataframe column names to names according to the naming convention that is recommended by Python
([*source*](https://stackoverflow.com/questions/41476150/removing-space-from-columns-in-pandas)).
By using this function on a dataframe:
1. All letters in the column names of the dataframe will be converted to lowercases
2. Both leading and trailing space of columns will be trimmed
3. All spaces will be replaced by snake_cases

```python
def column_names(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
#create dataframe to test if column_names(df) function is working
data = {'Name List':['Kristy', 'Jay'],
        'Age List':[20, 21]}
testing_df = pd.DataFrame(data)
print(testing_df)

column_names(testing_df)
print(testing_df)
```

```python
def subplot_boxplot(df, y_, col_list, width = 15, height = 15, titles = None, xlabels = None):
    nrows = int(np.ceil(len(col_list))) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows = nrows, ncols = 1, figsize = (width, height))
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for idx, ax in enumerate(ax): # Gives us an index value to get into all our lists
        sns.boxplot(x = col_list[idx], y = y_, data = df, ax = ax)
```

```python
#Subplot scatter function
def scatter_subplot(df, col_list, columns, t_list, xlabel_list, ylabel_list):
    nrows = int(np.ceil(len(col_list))) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows = nrows, ncols = 1, figsize = (20, 9)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.tight_layout()
    
    for i, ax in enumerate(ax): # Gives us an index value to get into all our lists
        ax.scatter(df[col_list[i]], df[columns[i]]) # need 2 sets of columns for scatter plots
        ax.set_title(t_list[i], fontsize = 15) #Set title for each subplot
        ax.set_xlabel(xlabel_list[i], fontsize = 12) #Set xlabel for each subplot
        ax.set_ylabel(ylabel_list[i], fontsize = 12) #Set ylabel for each subplot
```

```python
def heatmap(df, col_list):
    plt.figure(figsize = (10,6))
    corr = df[col_list].corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    sns.heatmap(corr, mask = mask, cmap = 'coolwarm', vmin = 0, vmax=1, annot = True);
```

```python
plt.figure(figsize=(8,5))
plt.scatter(ames['lot_frontage'], ames['saleprice'])
plt.title('Area of Lot Frontage vs Sale Price')
plt.xlabel('Lot Frontage Area')
plt.ylabel('Sale Price');
```

It is observed from the scatter chart that there are many outliers. Hence if we want to do imputation of NaN values on `lot_frontage`, we should use median instead of mean.

Also, since the area of each street connected to a property most likely have a similar area to other properties in its neighborhood , we will fill in NaN values by the median `lot_frontage` of the `neighborhood`.

```python
#Group by neighborhood and fill in NaN values with the median lot_frontage of the neighborhood
ames['lot_frontage'] = ames.groupby('neighborhood')['lot_frontage'].transform(lambda x: x.fillna(x.median()))
```

```python
ames.lot_frontage.isnull().sum()
```
