## Functions

#### column_name(df) 
Function to convert dataframe column names to names according to the naming convention that is recommended by Python
([*source*](https://stackoverflow.com/questions/41476150/removing-space-from-columns-in-pandas)).
By using this function on a dataframe:
1. All letters in the column names of the dataframe will be converted to lowercases
2. Both leading and trailing space of columns will be trimmed
3. All spaces will be replaced by snake_cases

```python
def column_names(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
def subplot_boxplot(df, y_, col_list, width = 15, height = 15, titles = None, xlabels = None):
    nrows = int(np.ceil(len(col_list))) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows = nrows, ncols = 1, figsize = (width, height))
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for idx, ax in enumerate(ax): # Gives us an index value to get into all our lists
        sns.boxplot(x = col_list[idx], y = y_, data = df, ax = ax)
```

```python
def scatter(df, x, y, title, xlabel, ylabel):
    fig, ax = plt.subplots()
    ax.scatter(df[x], df[y], s = 10)
    line = mlines.Line2D([0, 1], [0, 1], color = 'red')
    transform = ax.transAxes
    line.set_transform(transform)
    ax.add_line(line)
    ax.set_title(title, fontsize = 15) #Set title 
    ax.set_xlabel(xlabel, fontsize = 12) #Set xlabel 
    ax.set_ylabel(ylabel, fontsize = 12) #Set ylabel 
    plt.show()
```

```python
def scatter_subplots(df, col_list, columns, t_list, xlabel_list, ylabel_list):
    nrows = int(np.ceil(len(col_list))) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows = nrows, ncols = 1, figsize = (10, 9)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, ax in enumerate(ax): # Gives us an index value to get into all our lists
        ax.scatter(df[col_list[i]], df[columns[i]], s = 10) # need 2 sets of columns for scatter plots
        line = mlines.Line2D([0, 1], [0, 1], color = 'red')
        transform = ax.transAxes
        line.set_transform(transform)
        ax.add_line(line)
        ax.set_title(t_list[i], fontsize = 15) #Set title for each subplot
        ax.set_xlabel(xlabel_list[i], fontsize = 12) #Set xlabel for each subplot
        ax.set_ylabel(ylabel_list[i], fontsize = 12) #Set ylabel for each subplot
        
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9,  wspace=0.4, hspace=0.4)
```

```python
def heatmap(df, col_list):
    plt.figure(figsize = (10,6))
    corr = df[col_list].corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    sns.heatmap(corr, mask = mask, cmap = 'coolwarm', vmin = 0, vmax=1, annot = True);
```
