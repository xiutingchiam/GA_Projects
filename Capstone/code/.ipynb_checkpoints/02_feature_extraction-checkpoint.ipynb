## Create Signature Lists

### Train Dataset

```python
SIZE = 150

train_data = []
train_labels = []
train_data_names = []
```

```python
for per in os.listdir('../datasets/signatures_full/train/genuine/genuine_person'):
    for data in glob.glob('../datasets/signatures_full/train/genuine/genuine_person/'+per+'/*.*'):
        print(data)
        train_data_names.append(data)
        img = cv2.imread(data)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (SIZE,SIZE))
        train_data.append([img])
        train_labels.append(np.array(0))        
```

```python
for per in os.listdir('../datasets/signatures_full/train/forged/forged_person'):        
    for data in glob.glob('../datasets/signatures_full/train/forged/forged_person/'+per+'/*.*'):
        print(data)
        train_data_names.append(data)
        img = cv2.imread(data)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (SIZE,SIZE))
        train_data.append([img])
        train_labels.append(np.array(1))       
```

```python
train_data = np.array(train_data)/255.0
train_labels = np.array(train_labels)
```

### Test Dataset

```python
test_data = []
test_labels = []
test_data_names = []
```

```python
for per in os.listdir('../datasets/signatures_full/test/genuine/genuine_person'):
    for data in glob.glob('../datasets/signatures_full/test/genuine/genuine_person/'+per+'/*.*'):
        print(data)
        test_data_names.append(data)
        img = cv2.imread(data)
        test_data.append([img])
        test_labels.append(np.array(0))
```

```python
for per in os.listdir('../datasets/signatures_full/test/forged/forged_person'):
    for data in glob.glob('../datasets/signatures_full/test/forged/forged_person/'+per+'/*.*'):
        print(data)
        test_data_names.append(data)
        img = cv2.imread(data)
        test_data.append([img])
        test_labels.append(np.array(1))
```

```python
test_data = np.array(test_data)/255.0
test_labels = np.array(test_labels)
```

```python
with open('./train_data_names.pkl', 'wb') as fp:
    pickle.dump(train_data_names, fp)

with open('./test_data_names.pkl', 'wb') as fp:
    pickle.dump(test_data_names, fp)
```

```python
# Categorical labels
#print(train_labels)
train_labels = to_categorical(train_labels)
#print(train_data.shape)
# Reshaping
train_data = train_data.reshape(-1, SIZE,SIZE, 3)
test_data = test_data.reshape(-1, SIZE,SIZE, 3)
```

```python
test_data_names
```

## Models

```python
input_ = (150,150,3)
EPOCHS = 20
BS = 64
output_ = 2
```

### VGG16 Adam

```python
base_model = applications.VGG16(weights='imagenet', include_top=False, input_shape=input_)

model = Sequential()
data_augmentation = keras.Sequential([layers.experimental.preprocessing.RandomRotation(0.1)])
model.add(base_model)
model.add(Flatten(input_shape=base_model.output_shape[1:]))
model.add(Dense(256, activation='relu'))
model.add(Dense(output_, activation='softmax'))

model = Model(inputs=model.input, outputs=model.output)
model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=1e-4),
              metrics=['accuracy'])

model.summary()

earlyStopping = EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=3,
                              verbose=1)

early_stop=[earlyStopping]
progess = model.fit(train_data,train_labels, batch_size=BS,epochs=EPOCHS, callbacks=early_stop,validation_split=.3)
acc = progess.history['accuracy']
val_acc = progess.history['val_accuracy']
loss = progess.history['loss']
val_loss = progess.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.show()

plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()

intermediate_layer_model = Model(inputs=model.input,
                                 outputs=model.layers[-2].output)
intermediate_output_train = intermediate_layer_model.predict(train_data)
intermediate_output_test = intermediate_layer_model.predict(test_data)

np.save('./VGG16_Adam_train', intermediate_output_train)
np.save('./VGG16_Adam_test', intermediate_output_test)
```

```python
model.save_weights('../weights/vgg16-adam-model.h5')
```
